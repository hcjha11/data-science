```python
import plotly.graph_objects as go
import numpy as np
```

```python
x=np.outer(np.linspace(-2,2,30),np.ones(30))
y=x.copy().T
z=np.cos(x**2+y**2)

fig=go.Figure(data=[go.Surface(x=x,y=y,z=z)])
fig.show()
```
