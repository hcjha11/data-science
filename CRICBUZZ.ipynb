```python
from bs4 import BeautifulSoup
```

```python
import requests as request
```

```python
page=request.get("https://www.cricbuzz.com/cricket-stats/icc-rankings/men/batting")
page.content
```

```python
soup =BeautifulSoup(page.content,'html.parser')
```

```python
print(soup.prettify)
```

```python
players =soup.find_all('div',class_='cb-col cb-col-100 cb-padding-left0')
```

```python
print(len(players))
```

```python
for player in players:
  print(player.get_text())
```

```python
Position =[]
Player =[]
Rating =[]

for crikt in players:
  Positions =crikt.find('div',class_='cb-col cb-col-16 cb-rank-tbl cb-font-16').text
  Position.append(Position)
  Players  =crikt.find('div',class_='cb-col cb-col-67 cb-rank-plyr').text
  Player.append(Players)

  Ratings = crikt.find('div',class_='cb-col cb-col-17 cb-rank-tbl pull-right')
  Rating.append(Ratings)
```

```python
import pandas as pd
dic ={'Positions':Position,'Players':Player,'Ratings':Rating}
df =pd.DataFrame(dic)
```

```python
df.head()
```

```python
df.to_csv('Player.info.csv')
```
