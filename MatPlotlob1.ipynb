**Subplot**

```python
x =np.array([11,22,33])
y =np.array([13,23,45])


plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("hello")
plt.suptitle("program")


x =np.array([12,22,36])
y =np.array([14,33,46])

plt.subplot(1,2,2)
plt.plot(x,y)
plt.title('hello')
```

**Scatter**

```python
x =np.array([1,2,34,5,6,78,9,7])
y =np.array([1,2,3,45,67,89,7,9])

sizes= np.array([5,10,1,52,1,4,140,5])

plt.scatter(x,y,s =sizes,alpha=0.5)
plt.show
```

**bar graphs**

```python
x =np.array(['A','B','C','S','D','F','H','J'])
y =np.array([1,2,3,45,67,89,7,9])

plt.bar(x,y,color='red',width=0.5)
plt.show()
```

**Histogram**

```python
x =np.random.normal(140,10,250)

plt.hist(x)
plt.show()
```

**piechart**

```python
x =np.array([10,25,4,57,96,56,25])
label=['apple','orange','vanilla','red','green','brown','mango']
exp =[0.2,2,1,0,0,0,0]
plt.pie(x,labels=label,startangle=90,explode=exp,shadow=True)
plt.legend(title='Seven Fruits')
plt.show()
```
